```python
path = '/content/hansel_e_gretel.png'
image = Image.open(path)
display(image)
```

```python
!sudo apt install tesseract-ocr-ita  # Install Italian language data
```

```python
extracted_text = pytesseract.image_to_string(image)
extracted_text
```

```python
#extract text using a specific language (e.g., English)
language_text = pytesseract.image_to_string(image, lang='ita')
print(f"Extracted Text with Language Specification: {language_text}")
```

```python
!pip install googletrans==4.0.0-rc1
```

```python
text = language_text
from googletrans import Translator
translator = Translator()
translated_text = translator.translate(text, dest='en')
print(f"Translated Text: {translated_text.text}")
```
